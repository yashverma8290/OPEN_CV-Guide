```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass

cv2.namedWindow('myproject')

cv2.createTrackbar('th','myproject',0,255,nothing)

cv2.createTrackbar('lb','myproject',0,255,nothing)
cv2.createTrackbar('lg','myproject',0,255,nothing)
cv2.createTrackbar('lr','myproject',0,255,nothing)

cv2.createTrackbar('ub','myproject',255,255,nothing)
cv2.createTrackbar('ug','myproject',255,255,nothing)
cv2.createTrackbar('ur','myproject',255,255,nothing)

cap=cv2.VideoCapture(0)
while cap.isOpened():
         r,frame=cap.read()
         if r==True:


             thr=cv2.getTrackbarPos('th','myproject')

             LB=cv2.getTrackbarPos('lb','myproject')
             LG=cv2.getTrackbarPos('lg','myproject')
             LR=cv2.getTrackbarPos('lr','myproject')

             UB=cv2.getTrackbarPos('ub','myproject')
             UG=cv2.getTrackbarPos('ug','myproject')
             UR=cv2.getTrackbarPos('ur','myproject')



             lower=np.array([LB,LG,LR])
             upper=np.array([UB,UG,UR])
             






             
             frame=cv2.flip(frame,1)
             frame=cv2.resize(frame,(400,400))
             hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)


             msk=cv2.inRange(hsv,lower,upper)
             res=cv2.bitwise_and(frame,frame,mask=msk)
             fr=cv2.bitwise_not(res)

             ret,thi=cv2.threshold(msk,thr,255,cv2.THRESH_BINARY)
             cnt,hr=cv2.findContours(thi,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

             cv2.drawContours(frame,cnt,-1,(255,0,0),2)

             cv2.imshow('thr',thi)
             cv2.imshow('res',res)
             cv2.imshow('mask',msk)
             cv2.imshow('frame',frame)
             if cv2.waitKey(25) & 0xff==ord('p'):
                   break
         else:
             break














cap.release()
cv2.destroyAllWindows()
```
