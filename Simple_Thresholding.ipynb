# SIMPLE THRESHOLDINg

```python
img=cv2.imread('image17.png')
img=cv2.resize(img,(600,400))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

_,th=cv2.threshold(img,118,255,cv2.THRESH_BINARY)

cv2.imshow('threshold',th)
cv2.imshow('myimage',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Otsu's Binarization Threshold 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('image19.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist=cv2.calcHist([img],[0],None,[255],[0,255])
plt.plot(hist)
plt.show()

_,thres=cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imshow('myimage',img)
cv2.imshow('newimage',thres)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
