```python
import cv2
import numpy as np

# Load and resize the image
img = cv2.imread('image15.png')
img = cv2.resize(img, (500, 400))

# Callback function for trackbars (does nothing, just required)
def nothing(x):
    pass

# Create a window
cv2.namedWindow('Color Filter')

# Create trackbars for lower HSV values
cv2.createTrackbar('Lower Hue', 'Color Filter', 0, 179, nothing)
cv2.createTrackbar('Lower Sat', 'Color Filter', 0, 255, nothing)
cv2.createTrackbar('Lower Val', 'Color Filter', 0, 255, nothing)

# Create trackbars for upper HSV values
cv2.createTrackbar('Upper Hue', 'Color Filter', 0, 179, nothing)
cv2.createTrackbar('Upper Sat', 'Color Filter', 0, 255, nothing)
cv2.createTrackbar('Upper Val', 'Color Filter', 0, 255, nothing)

while True:
    # Convert image to HSV color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Get current positions of all trackbars
    lh = cv2.getTrackbarPos('Lower Hue', 'Color Filter')
    ls = cv2.getTrackbarPos('Lower Sat', 'Color Filter')
    lv = cv2.getTrackbarPos('Lower Val', 'Color Filter')

    uh = cv2.getTrackbarPos('Upper Hue', 'Color Filter')
    us = cv2.getTrackbarPos('Upper Sat', 'Color Filter')
    uv = cv2.getTrackbarPos('Upper Val', 'Color Filter')

    # Create lower and upper bounds for HSV
    lower_bound = np.array([lh, ls, lv])
    upper_bound = np.array([uh, us, uv])

    # Create mask for selected HSV range
    mask = cv2.inRange(hsv, lower_bound, upper_bound)

    # Apply mask to original image
    result = cv2.bitwise_and(img, img, mask=mask)

    # Show windows
    cv2.imshow('Original Image', img)
    cv2.imshow('HSV Image', hsv)
    cv2.imshow('Mask', mask)
    cv2.imshow('Filtered Result', result)

    # Press 'p' to exit
    if cv2.waitKey(1) & 0xFF == ord('p'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap=cv2.VideoCapture(0)

def nothing(x):
    pass

# Create a window
cv2.namedWindow('Color Filter')

# Create trackbars for lower HSV values
cv2.createTrackbar('Lower Hue', 'Color Filter', 0, 179, nothing)
cv2.createTrackbar('Lower Sat', 'Color Filter', 0, 255, nothing)
cv2.createTrackbar('Lower Val', 'Color Filter', 0, 255, nothing)

# Create trackbars for upper HSV values
cv2.createTrackbar('Upper Hue', 'Color Filter', 255, 179, nothing)
cv2.createTrackbar('Upper Sat', 'Color Filter', 255, 255, nothing)
cv2.createTrackbar('Upper Val', 'Color Filter', 255, 255, nothing)

while True:

    r,frame=cap.read()

    if r==True:
          img=cv2.resize(frame,(400,300))
          # Convert image to HSV color space
          hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

          # Get current positions of all trackbars
          lh = cv2.getTrackbarPos('Lower Hue', 'Color Filter')
          ls = cv2.getTrackbarPos('Lower Sat', 'Color Filter')
          lv = cv2.getTrackbarPos('Lower Val', 'Color Filter')

          uh = cv2.getTrackbarPos('Upper Hue', 'Color Filter')
          us = cv2.getTrackbarPos('Upper Sat', 'Color Filter')
          uv = cv2.getTrackbarPos('Upper Val', 'Color Filter')

          # Create lower and upper bounds for HSV
          lower_bound = np.array([lh, ls, lv])
          upper_bound = np.array([uh, us, uv])

          # Create mask for selected HSV range
          mask = cv2.inRange(hsv, lower_bound, upper_bound)

          # Apply mask to original image
          result = cv2.bitwise_and(img, img, mask=mask)

          # Show windows
          cv2.imshow('Original Image', img)
          cv2.imshow('HSV Image', hsv)
          cv2.imshow('Mask', mask)
          cv2.imshow('Filtered Result', result)

          # Press 'p' to exit
          if cv2.waitKey(1) & 0xFF == ord('p'):
              break

    else:
        break
cap.release()
cv2.destroyAllWindows()
```
