```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_w=np.ones((500,500,3),np.uint8)*255
blank_b=np.zeros((500,500,3),np.uint8)*255
```

```python
hist1=cv2.calcHist([blank_w],[0],None,[256],[0,256])
hist2=cv2.calcHist([blank_b],[0],None,[256],[0,256])

plt.plot(hist1)
plt.show()

plt.plot(hist2)
plt.show()

cv2.imshow('blank b',blank_b)
cv2.imshow('blank w',blank_w)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

def nothing(x):
    pass

# Load your image
img = cv2.imread('image17.png')  # <-- replace with your image name
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create a window
cv2.namedWindow('Black and White Adjuster')

# Create a trackbar for intensity control
cv2.createTrackbar('Intensity', 'Black and White Adjuster', 0, 255, nothing)

while True:
    # Get the current position of the trackbar
    intensity = cv2.getTrackbarPos('Intensity', 'Black and White Adjuster')

    # Apply threshold based on the intensity
    _, bw_img = cv2.threshold(img_gray, intensity, 255, cv2.THRESH_BINARY)

    # Show the image
    cv2.imshow('Black and White Adjuster', bw_img)

    # Break the loop when 'ESC' key is pressed
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))

img1=cv2.imread('image20.png')
img2=cv2.imread('image21.png')
img3=cv2.imread('image22.png')

hist1=cv2.calcHist([img1],[0],None,[256],[0,256])
hist2=cv2.calcHist([img2],[0],None,[256],[0,256])
hist3=cv2.calcHist([img3],[0],None,[256],[0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()

cv2.imshow('blank b1',img1)
cv2.imshow('blank b2',img2)
cv2.imshow('blank b3',img3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))

img1=cv2.imread('image3.jpg')
img1=cv2.resize(img1,(500,500))
b,g,r=cv2.split(img1)

hist1=cv2.calcHist([b],[0],None,[256],[0,256])
hist2=cv2.calcHist([g],[0],None,[256],[0,256])
hist3=cv2.calcHist([r],[0],None,[256],[0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()

cv2.imshow('blank b',b)
cv2.imshow('blank g',g)
cv2.imshow('blank r',r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(5,5))

img1=cv2.imread('image3.jpg')
img1=cv2.resize(img1,(500,500))
b,g,r=cv2.split(img1)

plt.hist(b.ravel(),255,(0,255))
plt.hist(g.ravel(),255,(0,255))
plt.hist(r.ravel(),255,(0,255))

cv2.imshow('blank b',b)
cv2.imshow('blank g',g)
cv2.imshow('blank r',r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
