```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('image23.png')
img=cv2.resize(img,(500,500))

gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist1=cv2.calcHist([gry],[0],None,[255],[0,255])
eq=cv2.equalizeHist(gry)
hist2=cv2.calcHist([eq],[0],None,[255],[0,255])

plt.plot(hist1)
plt.plot(hist2)
plt.show()

cv2.imshow('newimage',eq)
cv2.imshow('myimage',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread('image24.png')
img=cv2.resize(img,(500,500))

gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist1=cv2.calcHist([gry],[0],None,[255],[0,255])
eq=cv2.equalizeHist(gry)
hist2=cv2.calcHist([eq],[0],None,[255],[0,255])

cl=cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))
f=cl.apply(gry)

plt.plot(hist1)
plt.plot(hist2)
plt.plot(f,c='green')
plt.show()

cv2.imshow('refinedimage',f)
cv2.imshow('newimage',eq)
cv2.imshow('myimage',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
